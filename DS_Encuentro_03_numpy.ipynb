### Memoria y Velocidad

```python
import time
```

```python

%%time
L1 = range(100000)
L2 = range(100000)

A1 = np.arange(100000)
A2 = np.arange(100000)

s = time.time()
%time [a1+a2 for a1, a2 in zip(L1,L2)]
#print('python list:', time.time()-s)


s = time.time()
%time [L1[i]+L2[i] for i in range(len(L1))]
#print('python list:', time.time()-s)


s = time.time()
%time A1+A2
#print('numpy array:', time.time()-s)
```

* ¿Qué operaciones se pueden hacer tanto en un arreglo de Numpy como en una lista? Dar un ejemplo en una celda.

* ¿Cuál es la diferencia entre un arreglo de forma -shape- (n,), (n,1) y (1,n)? Pueden crear arreglos para intentar responder esa pregunta.

```python
#operaciones que se pueden realizar con una lista y un arreglo
a=[1,2,3,4]
b=np.arange(4)
print(a[0]) #indexado
print(b[0])

print(a[0:2]) # funcion corte
print(b[0:2])
```

```python
#operaciones se pueden hacer en un arreglo de Numpy pero NO en una lista
print(np.mean(b)) # se obtiene el promedio de un array
print(np.sum(b)) # suma de los elementos de un vector
```

**Ejercicio 4**:
* Crear un arreglo de ceros de `shape` (5,10).
* Reemplazar la segunda y cuarta fila con unos
* Reemplazar la tercera y octava columna con dos (2).

```python
arreglo1=np.zeros(50)
arreglo1.shape=(5,10)
arreglo1[1:2]=1  # reemplazo filas
arreglo1[3:4]=1
arreglo1[ :,2:3]=2   # reemplazo columnas
arreglo1[ :,7:8]=2
arreglo1
```
